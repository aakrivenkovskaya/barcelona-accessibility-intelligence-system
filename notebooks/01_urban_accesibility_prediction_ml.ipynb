### Data Vocabulary

1. **Gis_ID** (Internal Identifier): Unique identifier for each record in the dataset, used for internal tracking.

2. **N_Gis** (Incident Number): A unique number assigned to each reported incident, used for identifying and referencing specific incidents.

3. **X_ETRS89** (X Coordinate in ETRS89): The X coordinate of the location in the ETRS89 coordinate system, used for geographic mapping.

4. **Y_ETRS89** (Y Coordinate in ETRS89): The Y coordinate of the location in the ETRS89 coordinate system, used for geographic mapping.

5. **Longitud** (X Coordinate in WGS84): The X coordinate (longitude) of the location in the WGS84 coordinate system, often used in GPS systems.

6. **Latitud** (Y Coordinate in WGS84): The Y coordinate (latitude) of the location in the WGS84 coordinate system, often used in GPS systems.

7. **Incidència** (Incident Category): Describes the type of incident reported (e.g., pavement issues, slope problems, level changes).

8. **Alta** (Incident Report Date): The date when the incident was reported, used for tracking the timeliness and age of reports.

9. **Situació** (Situation - Segment, Intersection): Indicates whether the incident occurred in a street segment or at an intersection.

10. **Tipus_de_tram_carrer** (Street Segment Type): Categorizes the type of street segment (e.g., residential, commercial).

11. **Codi_Graf** (Graph Code of the City Council of Barcelona): A code representing the graphical mapping used by the City Council of Barcelona for urban planning.

12. **ID_element_inventari** (Inventory Element Code): Code identifying the specific inventory element related to the incident (e.g., a particular curb, ramp).

13. **Codi_Barri** (Neighborhood Code): Code representing the neighborhood where the incident occurred.

14. **Nom_Barri** (Neighborhood Name): The name of the neighborhood where the incident occurred.

15. **Codi_Districte** (District Code): Code representing the district where the incident occurred.

16. **Nom_Districte** (District Name): The name of the district where the incident occurred.

17. **Carrer** (Street Name): The name of the street where the incident occurred.

18. **Número_Postal** (Street Number): The street number of the location where the incident occurred.

19. **Ubicació** (Tactile Pavement Location - Correct/Incorrect): Indicates whether the tactile pavement location is correct or incorrect according to accessibility standards.

20. **Accesibilitat** (Accessibility - Yes/No): Indicates whether the area is accessible (Values: Yes, No), based on various criteria such as slope, pavement condition, and obstacles.

21. **Tipus** (Level Change Type): Describes the type of level change present (e.g., steps, ramps, mechanical stairs, elevators).

22. **Quantitat** (Quantity of Similar Obstacles): Number of similar obstacles present in the area, providing a count of repeated issues.

23. **Mesura** (Street Width): The width of the street, which can affect accessibility, especially for those with mobility aids.

24. **Amplada_lliure_pas_central** (Central Free Passage Width): The width of the free passage available in the central part of the street.

25. **Amplada_lliure_pas_façana** (Façade Free Passage Width): The width of the free passage available along the façade of buildings.

26. **Adossat_a_façana_<0.50m** (Obstacles Attached to Façade <0.50m): Indicates whether there are obstacles attached to the façade within 0.50 meters, which can impede accessibility.

27. **Llargada** (Ramp Length): The length of the ramp, important for determining the ramp's compliance with accessibility standards.

28. **Amplada** (Level Change Width): The width of the level change (e.g., steps or ramps).

29. **Amplada_Gual** (Curb Width): The width of the curb cut or curb ramp.

30. **Amplada_Pas_Vianants** (Pedestrian Crossing Width): The width of the pedestrian crossing.

31. **Tipus_pas_vianants** (Pedestrian Crossing Type): The type of pedestrian crossing, which can affect safety and accessibility.

32. **Alçada** (Tactile Pavement Relief Height): The height of the tactile pavement relief, which helps visually impaired individuals navigate.

33. **Amidament** (Tactile Pavement Length): The length of the tactile pavement, providing guidance for visually impaired individuals.

34. **Mesura_mínima** (Minimum Free Passage Width): The minimum width of the free passage available, critical for accessibility.

35. **Pendent_Longitudinal** (Longitudinal Slope): The longitudinal slope of the street segment, affecting mobility for wheelchairs and other aids.

36. **Pendent_Transversal** (Transversal Slope): The transversal slope of the street segment, affecting stability and ease of movement.

37. **Pendent** (Curb Slope): The slope of the curb ramp, important for safe and accessible curb transitions.

38. **Graó_Desnivell** (Curb Height Difference): The height difference at curbs, which can be a barrier for wheelchair users.

39. **Tipus_Direccional** (Directional Tactile Pavement Width): The width of directional tactile pavement, used to guide visually impaired pedestrians.

40. **Tipus_Botons** (Warning Tactile Pavement Width): The width of warning tactile pavement, used to signal hazards to visually impaired pedestrians.

41. **Falta_Direccional** (Missing Directional Tactile Pavement): Indicates if the directional tactile pavement is missing, which can pose a hazard.

42. **Falta Botons** (Missing Warning Tactile Pavement): Indicates if the warning tactile pavement is missing, which can pose a hazard.

43. **Detall_Altres** (Additional Details): Provides additional clarifications or details about the incident or location.

44. **Profunditat** (Ramp Depth): The depth dimensions of the ramp, which is important for accessibility compliance.

45. **Cas_No_Compleix_Mesura_Longitudinal** (Non-compliance with Longitudinal Measure): Indicates if there is non-compliance with the longitudinal slope measure.

46. **Cas_No_Compleix_Mesura_Transversal** (Non-compliance with Transversal Measure): Indicates if there is non-compliance with the transversal slope measure.

47. **Senyalització_Horitzontal** (Horizontal Signage for Reserved Parking Spaces): Indicates the presence of horizontal signage for reserved parking spaces, which is necessary for proper identification.

48. **Senyalització_Vertical** (Vertical Signage for Reserved Parking Spaces): Indicates the presence of vertical signage for reserved parking spaces, which is necessary for proper identification.

49. **Reserva_Accessible** (Accessibility of Reserved Parking Spaces): Indicates whether the reserved parking spaces are accessible, ensuring compliance with accessibility standards.

#### 4. Outlier Handing

Getting back to the problem of highly imbalanced distribution of numerical columns, let's start dealing with outliers before scaling.
After multiple trials was found out that our dataset is super sensitive for outlier handling approaches, for this reason Soft Outlier Handling strategy was selected. Those steps will be taken:

- **Winsorization**: Applied to handle extreme outliers.
- **Log Transformation**: Applied to reduce skewness.

```python
data[numerical_cols].describe()
```

```python
from scipy.stats import mstats

# Identify columns that are more sensitive for outliers handling
problematic_cols = ['Llargada', 'Amplada', 'Cas No Compleix Mesura Transv.']

# Apply winsorization with adjusted thresholds
for col in numerical_cols:
    if col in problematic_cols:
        # Apply a less aggressive winsorization
        data[col] = mstats.winsorize(data[col], limits=[0.001, 0.001])
    else:
        # Default winsorization for other columns
        data[col] = mstats.winsorize(data[col], limits=[0.01, 0.01])

# Verify if problematic columns still have zero variance
zero_variance_cols = [col for col in numerical_cols if data[col].std() == 0]
print("Columns with zero variance after adjusted winsorization:", zero_variance_cols)

data[numerical_cols].describe()
```

```python
# Log Transformation for Skewed Columns

# Identify skewed columns
skewness = data[numerical_cols].apply(lambda x: x.skew()).abs()
skewed_cols = skewness[skewness > 0.75].index

# Apply log transformation to skewed columns
data[skewed_cols] = data[skewed_cols].apply(lambda x: np.log1p(x))

# Describe the transformed skewed columns to verify
data[skewed_cols].describe()
```

It seems that the Log transformation was applied correctly. The Log transformation is generally used to compress the range of skewed data and bring extreme values closer to the center. Here are some observations from obtained results:

**Quantitat**:

- Mean: 0.750890
- Std: 0.254674
- Min: 0.693147
- Max: 2.397895

This indicates that the Quantitat values have been compressed, and the range is now much narrower compared to the original values.

**Mesura**:

- Mean: 5.695154

- Min: 4.709530
- Max: 6.398595
  
Similarly, the Mesura column values have been compressed.

Other Columns:

**Amidament**: Shows a compressed range after log transformation.

**Profunditat**: Shows values compressed between 4.615121 and 5.198497.

The mean values are closer to the median, and the standard deviation has been reduced, which indicates that the data distribution has become more normal.

#### Define Models and Parameter Distributions

```python
models = {
    'LogisticRegression': LogisticRegression(max_iter=2000),
    'RandomForestClassifier': RandomForestClassifier(),
    'GradientBoostingClassifier': GradientBoostingClassifier()
}

param_distributions = {
    'LogisticRegression': {
        'C': np.logspace(-4, 4, 20),
        'solver': ['liblinear', 'saga']
    },
    'RandomForestClassifier': {
        'n_estimators': [100, 200, 300],
        'max_depth': [None, 10, 20, 30],
        'min_samples_split': [2, 5, 10]
    },
    'GradientBoostingClassifier': {
        'n_estimators': [100, 200, 300],
        'learning_rate': [0.01, 0.1, 0.2],
        'max_depth': [3, 4, 5]
    }
}
```

#### Function to Perform Randomized Search and Cross-Validation

For the sake of the faster time execution, we will use Randomized Search instead of Grid Search. Additionally, the 3-fold cross-validation will be applied to evaluate the performance of different parameter settings.

```python
def perform_random_search(model, param_distributions, X_train, y_train):
    random_search = RandomizedSearchCV(
        model,
        param_distributions=param_distributions,
        n_iter=50,  # Number of parameter settings that are sampled
        cv=3,  # Cross-validation splitting strategy
        verbose=1,  # Control the verbosity
        n_jobs=-1,  # Number of jobs to run in parallel
        scoring='accuracy',  # Evaluation metric
        random_state=42  # Ensure reproducibility
    )
    random_search.fit(X_train, y_train)
    return random_search.best_estimator_, random_search.best_params_, random_search.best_score_
```

```python
best_models = {}
for model_name in models.keys():
    print(f"Performing Randomized Search for {model_name}")
    model = models[model_name]
    param_dist = param_distributions[model_name]
    best_model, best_params, best_score = perform_random_search(model, param_dist, X_train, y_train)
    best_models[model_name] = {
        'best_model': best_model,
        'best_params': best_params,
        'best_score': best_score
    }
    print(f"Best parameters for {model_name}: {best_params}")
    print(f"Best cross-validation score for {model_name}: {best_score}")
    print("===========================================\n")
```
